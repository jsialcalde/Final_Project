```python
import pandas as pd
```

```python
# import cleaned up movie data
movieDF = pd.read_csv('Inflation_Adjusted_Movie.csv')
movieDF.dtypes
```

```python
# import economic indicator data
econDF = pd.read_csv('USA_Multiyear_Indicators.csv')
econDF.dtypes
```

```python
econDF.head()
```

```python
# rename TIME
econDF=econDF.rename(columns={'TIME':'Year'})

# create rolling average columns
econDF['GDP_AVG']=(econDF.GDP+econDF.GDP_1+econDF.GDP_2)/3
econDF['Unemployment_AVG']=(econDF.Unemployment+econDF.Unemployment_1+econDF.Unemployment_2)/3
econDF['Confidence_AVG']=(econDF.Confidence+econDF.Confidence_1+econDF.Confidence_2)/3
econDF['Median_Income_AVG']=(econDF.Median_Income+econDF.Median_Income_1+econDF.Median_Income_2)/3
econDF.head()
```

```python
# join on year
mergedDF = movieDF.merge(econDF, on='Year',how='left')
```

```python
mergedDF.head(20)
```

```python
mergedDF.columns
```

```python
# drop any unnecessary columns
mergedDF = mergedDF[['Title','Year','Release Date', 'Genre', 'IMDb Rating',
       'Metascore', 'Rated', 'Box_Office_Adjusted', 
       'GDP', 'GDP_1', 'GDP_2', 'Unemployment', 'Unemployment_1',
       'Unemployment_2', 'Confidence', 'Confidence_1', 'Confidence_2',
       'Median_Income', 'Median_Income_1', 'Median_Income_2', 'GDP_AVG',
       'Unemployment_AVG', 'Confidence_AVG', 'Median_Income_AVG']]
# export to csv
mergedDF.to_csv('mergedData.csv')
```
