```python
import pandas as pd

# from collections import Counter
# from string import punctuation

import nltk
# needed only if nltk isn't already installed
# nltk.download()
# from nltk.corpus import stopwords
# from nltk import word_tokenize
```

```python
# import cleaned up movie data
movieDF = pd.read_csv('Inflation_Adjusted_Movie.csv')
movieDF.dtypes
```

```python
# import economic indicator data
econDF = pd.read_csv('USA_Multiyear_Indicators.csv')
econDF.dtypes
```

```python
econDF.head()
```

```python
# rename TIME
econDF=econDF.rename(columns={'TIME':'Year'})

# create rolling average columns
econDF['GDP_AVG']=(econDF.GDP+econDF.GDP_1+econDF.GDP_2)/3
econDF['Unemployment_AVG']=(econDF.Unemployment+econDF.Unemployment_1+econDF.Unemployment_2)/3
econDF['Confidence_AVG']=(econDF.Confidence+econDF.Confidence_1+econDF.Confidence_2)/3
econDF['Median_Income_AVG']=(econDF.Median_Income+econDF.Median_Income_1+econDF.Median_Income_2)/3
econDF.head()
```

```python
# join on year
mergedDF = movieDF.merge(econDF, on='Year',how='left')
```

```python
mergedDF.head(20)
```

```python
mergedDF.columns
```

```python
# drop any unnecessary columns
mergedDF = mergedDF[['Title','Year','Release Date', 'Genre', 'IMDb Rating',
       'Metascore', 'Rated', 'Box_Office_Adjusted', 
       'GDP', 'GDP_1', 'GDP_2', 'Unemployment', 'Unemployment_1',
       'Unemployment_2', 'Confidence', 'Confidence_1', 'Confidence_2',
       'Median_Income', 'Median_Income_1', 'Median_Income_2', 'GDP_AVG',
       'Unemployment_AVG', 'Confidence_AVG', 'Median_Income_AVG']]
# export to csv
mergedDF.to_csv('mergedData.csv')
```

```python
mergedDF.Genre.unique()
```

```python
newDF = mergedDF
newDF.head()
```

```python
top_N = 30
a = newDF.Genre.str.lower().str.cat(sep=' ')
words = nltk.tokenize.word_tokenize(a)
word_dist = nltk.FreqDist(words)
print (word_dist)
```

```python
rslt = pd.DataFrame(word_dist.most_common(top_N),
                    columns=['Word', 'Frequency'])
print(rslt)
```

```python
rslt = rslt.iloc[1:23]
```

```python
rslt
```

```python
rsltList = list(rslt.Word)
rsltList
```

```python
# new data frame with split value columns 
genreDF = newDF["Genre"].str.split(" ", n = 20, expand = True) 
genreDF.head()
```

```python
# lowercase all genres and replace ','
for index in (range(genreDF.shape[1])):
    genreDF[index]=genreDF[index].str.lower().str.replace(',','')
```

```python
# create placeholder columns in new DF to count category
# genreDFCount = pd.DataFrame()
for genre in rsltList:
    genreDF[genre] = 0
genreDF.head()
```

```python
# loop through every row of genreDF
for key,value in genreDF.iterrows():
# loop through every column of genreDF

    for genre in value[0:7]:
        if genre in rsltList:
#             print(genreDF[genre][key])
            genreDF[genre][key] = 1
#             print(genreDF[genre])
    
            
```

```python
# drop columns 0-7
genreDF = genreDF.iloc[:,8:]
```

```python
# merge genreDF back with mergedDF, join on index 
mergedDF = mergedDF.join(other=genreDF,how='left')
```
