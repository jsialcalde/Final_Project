```python
import pandas as pd
```

```python
oecd_df = pd.read_csv("OECD_Indicators.csv", index_col=0)
oecd_df.head()
```

```python
gdp_1_list = [0, 0]
gdp_2_list = [0, 0]
unemploy_1_list = [0, 0]
unemploy_2_list = [0, 0]
conf_1_list = [0, 0]
conf_2_list = [0, 0]

for index, row in oecd_df.iterrows():
        if index >= 2:
            gdp_1 = oecd_df.loc[index - 1, "GDP"]
            gdp_2 = oecd_df.loc[index - 2, "GDP"]
            gdp_1_list.append(gdp_1)
            gdp_2_list.append(gdp_2)
            
            unemploy_1 = oecd_df.loc[index - 1, "Unemployment"]
            unemploy_2 = oecd_df.loc[index - 2, "Unemployment"]
            unemploy_1_list.append(unemploy_1)
            unemploy_2_list.append(unemploy_2)
            
            conf_1 = oecd_df.loc[index - 1, "Confidence"]
            conf_2 = oecd_df.loc[index - 2, "Confidence"]
            conf_1_list.append(conf_1)
            conf_2_list.append(conf_2)
```

```python
oecd_full_df = oecd_df.copy()

gdp_idx = oecd_full_df.columns.get_loc("GDP")
oecd_full_df.insert(gdp_idx + 1, "GDP_1", gdp_1_list)
oecd_full_df.insert(gdp_idx + 2, "GDP_2", gdp_2_list)

unemploy_idx = oecd_full_df.columns.get_loc("Unemployment")
oecd_full_df.insert(unemploy_idx + 1, "Unemployment_1", unemploy_1_list)
oecd_full_df.insert(unemploy_idx + 2, "Unemployment_2", unemploy_2_list)

conf_idx = oecd_full_df.columns.get_loc("Confidence")
oecd_full_df.insert(conf_idx + 1, "Confidence_1", conf_1_list)
oecd_full_df.insert(conf_idx + 2, "Confidence_2", conf_2_list)

oecd_full_df = oecd_full_df[oecd_full_df["TIME"] >= 1992]
oecd_full_df = oecd_full_df.reset_index(drop=True)

oecd_full_df
```

```python
median_df = pd.read_csv("USA_Median_Income.csv")
median_df
```

```python
median_1_list = [0, 0]
median_2_list = [0, 0]

for index, row in median_df.iterrows():
        if index >= 2:
            median_1 = median_df.loc[index - 1, "Median_Income"]
            median_2 = median_df.loc[index - 2, "Median_Income"]
            median_1_list.append(median_1)
            median_2_list.append(median_2)
```

```python
median_full_df = median_df.copy()

median_idx = median_full_df.columns.get_loc("Median_Income")
median_full_df.insert(median_idx + 1, "Median_Income_1", median_1_list)
median_full_df.insert(median_idx + 2, "Median_Income_2", median_2_list)

median_full_df = median_full_df[median_full_df["TIME"] >= 1992]
median_full_df = median_full_df.reset_index(drop=True)

median_full_df
```

```python
oecd_full_df = pd.merge(oecd_full_df, median_full_df, on="TIME")
oecd_full_df.head()
```

```python
oecd_full_df.to_csv("USA_Multiyear_Indicators.csv")
```
