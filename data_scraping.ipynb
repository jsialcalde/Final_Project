```python
import pandas as pd
```

```python
gdp_df = pd.read_csv("OECD_GDP_1990-2017.csv")
gdp_df.head()
```

```python
gdp_usa_df = gdp_df[gdp_df["LOCATION"] == "USA"]
gdp_usa_df = gdp_usa_df.reset_index(drop=True)
gdp_usa_df = gdp_usa_df.rename(columns={"Value": "GDP"})
gdp_usa_df = gdp_usa_df[["LOCATION", "TIME", "GDP"]]
gdp_usa_df
```

```python
unemploy_df = pd.read_csv("OECD_Unemployment_1990-2018.csv")
unemploy_df.head()
```

```python
unemploy_usa_df = unemploy_df[unemploy_df["LOCATION"] == "USA"]
unemploy_usa_df = unemploy_usa_df.reset_index(drop=True)
unemploy_usa_df = unemploy_usa_df.rename(columns={"Value": "Unemployment"})
unemploy_usa_df = unemploy_usa_df[["LOCATION", "TIME", "Unemployment"]]
unemploy_usa_df
```

```python
consumer_df = pd.read_csv("OECD_Consumer_Confidence-1990-2019.csv")
consumer_df.head()
```

```python
consumer_usa_df = consumer_df[consumer_df["LOCATION"] == "USA"]
consumer_usa_df = consumer_usa_df.reset_index(drop=True)
consumer_usa_df = consumer_usa_df.rename(columns={"Value": "Confidence"})
consumer_usa_df = consumer_usa_df[["LOCATION", "YEAR", "MONTH", "Confidence"]]
consumer_usa_df.head()
```

```python
cyearly_df = consumer_usa_df.groupby("YEAR").mean()
cyearly_df = cyearly_df.drop("MONTH", axis=1)
cyearly_df = cyearly_df.reset_index()
cyearly_df = cyearly_df.rename(columns={"YEAR": "TIME"})
cyearly_df.insert(0, "LOCATION", "USA")
cyearly_df
```

```python
all_df = pd.merge(gdp_usa_df, unemploy_usa_df, on=["LOCATION", "TIME"])
all_df = pd.merge(all_df, cyearly_df, on=["LOCATION", "TIME"])
all_df.head()
```

```python
all_df.to_csv("OECD_Indicators.csv")
```
